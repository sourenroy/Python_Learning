```python
# Sum of odd numbers from 1 to 50
total = 0
for i in range(1, 51):
    if i % 2 != 0:
        total += i
print("Sum of odd numbers:", total)
```

```python
# Leap Year
year = 2024
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
# Factorial of a number
num = 5
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial of", num, "is", fact)
```

Reversed Number

```python
# Reversed Number
num = 12345
reversed_num = int(str(num)[::-1])
print("Reversed Number:", reversed_num)
```

Armstrong Number

```python
# Armstrong Number
num = 153
order = len(str(num))
sum = sum(int(digit) ** order for digit in str(num))

if num == sum:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

Palindrome Number

```python
# Palindrome Number
num = 121
if str(num) == str(num)[::-1]:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

Prime Number

```python
# Prime Number
num = 29
if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
else:
    print("Not a Prime Number")
```

Pattern

```python
rows = 5
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=' ')
    print()
```

```python
rows = 5
for i in range(rows, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=' ')
    print()
```

```python
rows = 5
for i in range(1, rows + 1):
    print('  ' * (rows - i), end='')  # Print leading spaces
    for j in range(1, i + 1):
        print(j, end=' ')
    print()
```

```python
#  Star Pyramid
for i in range(1, 6):
    print(' ' * (5 - i) + '* ' * i)
```

```python
# Diamond Pattern
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + '* ' * i)
for i in range(n - 1, 0, -1):
    print(' ' * (n - i) + '* ' * i)
```

```python
# String operation
s = "helloWorld"

# Uppercase
print(s.upper())  # HELLOWORLD

# Lowercase
print(s.lower())  # helloworld

# Capitalize
print(s.capitalize())  # Helloworld

# Reverse
print(s[::-1])  # dlroWolleh

# Length
print(len(s))  # 10

# Palindrome Check
print("Palindrome" if s == s[::-1] else "Not Palindrome")

# Slicing
print(s[0:5])  # hello
print(s[-5:])  # World
```

```python
# List Operation
import itertools

lst = [1, 2, 3]

# Append
lst.append(4)
print(lst)  # [1, 2, 3, 4]

# Delete
del lst[1]
print(lst)  # [1, 3, 4]

# Slicing
print(lst[1:])  # [3, 4]

# Permutations
perms = list(itertools.permutations(lst))
print(perms)
```
