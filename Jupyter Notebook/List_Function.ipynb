```python
# Create a 2D list (matrix)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Traverse the 2D list
print("Traversing the 2D list:")
for row in matrix:
    for item in row:
        print(item, end=' ')
    print()
```

```python
# Function to rotate a list by k positions
def rotate_list(lst, k):
    k = k % len(lst)  # Handle cases where k > length
    return lst[-k:] + lst[:-k]

# Example usage
original = [1, 2, 3, 4, 5]
rotated = rotate_list(original, 2)
print("Original list:", original)
print("Rotated list by 2 positions:", rotated)
```

```python
# Function to check if two lists are permutations
def are_permutations(lst1, lst2):
    return sorted(lst1) == sorted(lst2)

# Example usage
print("Are [1,2,3] and [3,2,1] permutations?", are_permutations([1,2,3], [3,2,1]))
```

```python
# Function to find common elements
def common_elements(lst1, lst2):
    return list(set(lst1) & set(lst2))

# Example usage
print("Common elements:", common_elements([1,2,3,4], [3,4,5,6]))
```

```python
# Function to generate Pascal's Triangle
def pascal_triangle(n):
    triangle = []
    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
        triangle.append(row)
    return triangle

# Example usage
print("Pascal's Triangle (5 rows):")
for row in pascal_triangle(5):
    print(row)
```

```python
# Convert list to tuple
lst = [10, 20, 30]
tup = tuple(lst)
print("Converted tuple:", tup)
```

```python
# Function to find second largest element
def second_largest(lst):
    unique = list(set(lst))  # Remove duplicates
    unique.sort()
    return unique[-2] if len(unique) >= 2 else None

# Example usage
print("Second largest in [10, 20, 30, 20]:", second_largest([10, 20, 30, 20]))
```

```python
# Function to transpose a matrix
def transpose(matrix):
    return [list(row) for row in zip(*matrix)]

# Example usage
matrix = [[1, 2, 3], [4, 5, 6]]
transposed = transpose(matrix)
print("Original matrix:")
for row in matrix:
    print(row)
print("Transposed matrix:")
for row in transposed:
    print(row)
```
